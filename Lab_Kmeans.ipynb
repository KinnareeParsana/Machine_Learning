```
import random
```

```
init_centroids = random.sample(range(0,len(df)),4)
init_centroids
```

```
centroids=[]
for i in init_centroids:
    centroids.append(df.iloc[i])

print(centroids)
```

```
centroids = np.array(centroids)
centroids
```

```
def calc_distance(X1,X2):
    return ((sum(X1-X2)**2)**0.5)
```

```
def findclosestcentroid(c,X):
    assign_centroids=[]
    for i in X:
        distance_list=[]
        for j in c:
            distance_list.append(calc_distance(i,j))
        assign_centroids.append(np.argmin(distance_list))
    return assign_centroids
```

```
get_centroid = findclosestcentroid(centroids,X)
```

```
def calc_centroids(cluster,X):
    new_centroid = []
    new_df = pd.concat([pd.DataFrame(X),pd.DataFrame(cluster,columns=['clusters'])],axis=1)
    for c in set(new_df['clusters']):
        current_cluster = new_df[new_df['clusters']==c][new_df.columns[:-1]] 
        current_mean = current_cluster.mean(axis=0)
        new_centroid.append(current_mean)
    return new_centroid
```

```
for i in range(30):
    get_centroid = findclosestcentroid(centroids,X)     
    centroids = calc_centroids(get_centroid,X)

    plt.figure()
    plt.scatter(np.array(centroids)[:,0],np.array(centroids)[:,1],color='black')    
    plt.scatter(X[:,0],X[:,1],color='blue',alpha=0.5) 
    plt.show()
```
